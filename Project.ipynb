```python
import csv
year_data = {}
for i in range(2012, 2017) :
    if i < 2014 :
        year_data[i] = []
        j = open(str(i) + '년_범죄자_직업.csv')
        data_i = csv.reader(j)
        data_i = list(data_i)
        j.close()
        if i == 2012 :
            for row in data_i :
                if '계' in row[0] :
                    year_data[i].append(int(row[3]))
                    year_data[i].append(int(row[22]))
                    year_data[i].append(int(row[14+26]))
                    year_data[i].append(int(row[23+26]))
                    break
        else :
            for row in data_i :
                if '계' in row[0] :
                    year_data[i].append(int(row[3]))
                    year_data[i].append(int(row[20]))
                    year_data[i].append(int(row[10+26]))
                    year_data[i].append(int(row[19+26]))
                    break

    else :
        k = open(str(i) + '년_범죄자_직업.csv')
        data_i = csv.reader(k)
        j.close()
        next(data_i)
        data_i = list(data_i)
        자영업_i=0
        피고용자_i=0
        전문직_i=0
        기타_i=0
        if i == 2014 :
            for row in data_i :
                if row[0] != "":
                    자영업_i = 자영업_i + int(row[3])
                    피고용자_i = 피고용자_i + int(row[20])
                    전문직_i = 전문직_i + int(row[36])
                    기타_i = 기타_i + int(row[45])
            year_data[i] = [자영업_i, 피고용자_i, 전문직_i, 기타_i]

        elif i == 2015 :
            for row in data_i :
                if row[0] != "":
                    자영업_i = 자영업_i + int(row[2])
                    피고용자_i = 피고용자_i + int(row[19])
                    전문직_i = 전문직_i + int(row[26+9])
                    기타_i = 기타_i + int(row[26+18])
            year_data[i] = [자영업_i, 피고용자_i, 전문직_i, 기타_i]

        else :
            for row in data_i :
                if row[0] != "":
                    자영업_i = 자영업_i + int(row[3])
                    피고용자_i = 피고용자_i + int(row[20])
                    전문직_i = 전문직_i + int(row[36])
                    기타_i = 기타_i + int(row[45])
            year_data[i] = [자영업_i, 피고용자_i, 전문직_i, 기타_i]


import matplotlib.pyplot as plt
plt.rc('font', family = 'Malgun Gothic')
print('연도 [자영업, 피고용자, 전문직, 기타]')
for year in year_data :
    print(year, year_data[year])
    plt.pie(year_data[year], labels = ['자영업', '피고용자', '전문직', '기타'])
    plt.axis('equal')
    plt.title(str(year) + '년 직종별 범죄자 수')
    plt.show()

plt.plot([year_data[2012][0], year_data[2013][0], year_data[2014][0], year_data[2015][0], year_data[2016][0]], label = '자영업')
plt.plot([year_data[2012][1], year_data[2013][1], year_data[2014][1], year_data[2015][1], year_data[2016][1]], label = '피고용자')
plt.plot([year_data[2012][2], year_data[2013][2], year_data[2014][2], year_data[2015][2], year_data[2016][2]], label = '전문직')
plt.plot([year_data[2012][3], year_data[2013][3], year_data[2014][3], year_data[2015][3], year_data[2016][3]], label = '기타')


plt.title('연도에 따른 직종별 범죄 수 그래프') 

plt.xticks(range(5),['2012년','2013년','2014년','2015년', '2016년']) 

plt.xlabel('연도') 
plt.ylabel('범죄수') 

plt.show() 
```
